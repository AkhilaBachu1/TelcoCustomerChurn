```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("TelcoCustomerChurnClean.csv")
```

```python
dataset.head()
```

```python
sns.histplot(dataset['tenure'])
title = 'Distribution : ' + 'tenure'
plt.title(title)
```

```python
sns.histplot(dataset['MonthlyCharges'])
title = 'Distribution : ' + 'MonthlyCharges'
plt.title(title)
```

```python
sns.histplot(dataset['TotalCharges'])
title = 'Distribution : ' + 'TotalCharges'
plt.title(title)
```

```python
corr_matrix = dataset.corr()
fig, ax = plt.subplots(figsize=(16, 16))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.countplot(x="tenure", hue="Churn", data=dataset)
plt.legend(["No Churn", 'churn'])
plt.show()
```

```python
fig = plt.subplots(figsize = (8,8))
ax = sns.boxplot(x = 'gender',y = 'tenure',data = dataset,hue = 'Churn');
plt.title('tenure vs ' + 'gender')
```

```python
fig = plt.subplots(figsize = (8,8))
ax = sns.boxplot(x = 'SeniorCitizen',y = 'tenure',data = dataset,hue = 'Churn');
plt.title('tenure vs ' + 'SeniorCitizen')
```

```python
fig = plt.subplots(figsize = (8,8))
ax = sns.boxplot(x = 'Partner',y = 'tenure',data = dataset,hue = 'Churn');
plt.title('tenure vs ' + 'Partner')
```
